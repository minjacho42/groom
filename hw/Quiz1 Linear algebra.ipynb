6. Vector $x \in R^2$에 대한 Transformation $T_1$: (1) x축에 대해 reflection 후 (2) y축에 대해 reflection. \\
 Vector $x \in R^2$에 대한 Transformatiopn $T_2$: 시계방향으로 180도 회전. \\
 이때 $T_1$은 $T_2$와 같다. \\

A: True

x축에 대한 reflection을 시키는 행렬을 $X$ y축에 대한 reflection을 시키는 행렬을 $Y$
시계방향으로 180도 회전을 시키는 행렬을 $Z$ 라고 하자.

이때 $T1$ 은 좌표계 A에 $(YX)A$ 를 하면 되고

$T2$ 는 좌표계 A에 $ZA$ 를 하면 된다.

이때 $Y = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}$  $X = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$ $YX = \begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix}$

$Z = \begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix}$ 이기 때문에 $T1 == T2$ 이다
