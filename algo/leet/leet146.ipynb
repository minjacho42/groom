<h2>146.LRU Cache</h2>

<div><p>Design a data structure that follows the constraints of a <strong><a href="https://en.wikipedia.org/wiki/Cache_replacement_policies#LRU" target="_blank">Least Recently Used (LRU) cache</a></strong>.</p>

<p>Implement the <code>LRUCache</code> class:</p>

<ul>
	<li><code>LRUCache(int capacity)</code> Initialize the LRU cache with <strong>positive</strong> size <code>capacity</code>.</li>
	<li><code>int get(int key)</code> Return the value of the <code>key</code> if the key exists, otherwise return <code>-1</code>.</li>
	<li><code>void put(int key, int value)</code> Update the value of the <code>key</code> if the <code>key</code> exists. Otherwise, add the <code>key-value</code> pair to the cache. If the number of keys exceeds the <code>capacity</code> from this operation, <strong>evict</strong> the least recently used key.</li>
</ul>

<p>The functions <code>get</code> and <code>put</code> must each run in <code>O(1)</code> average time complexity.</p>

<p>&nbsp;</p>
<p><strong>Example 1:</strong></p>

<pre><strong>Input</strong>
["LRUCache", "put", "put", "get", "put", "get", "put", "get", "get", "get"]
[[2], [1, 1], [2, 2], [1], [3, 3], [2], [4, 4], [1], [3], [4]]
<strong>Output</strong>
[null, null, null, 1, null, -1, null, -1, 3, 4]

<strong>Explanation</strong>
LRUCache lRUCache = new LRUCache(2);
lRUCache.put(1, 1); // cache is {1=1}
lRUCache.put(2, 2); // cache is {1=1, 2=2}
lRUCache.get(1);    // return 1
lRUCache.put(3, 3); // LRU key was 2, evicts key 2, cache is {1=1, 3=3}
lRUCache.get(2);    // returns -1 (not found)
lRUCache.put(4, 4); // LRU key was 1, evicts key 1, cache is {4=4, 3=3}
lRUCache.get(1);    // return -1 (not found)
lRUCache.get(3);    // return 3
lRUCache.get(4);    // return 4
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= capacity &lt;= 3000</code></li>
	<li><code>0 &lt;= key &lt;= 10<sup>4</sup></code></li>
	<li><code>0 &lt;= value &lt;= 10<sup>5</sup></code></li>
	<li>At most <code>2 * 10<sup>5</sup></code> calls will be made to <code>get</code> and <code>put</code>.</li>
</ul>
</div>

```python
"""
Runtime: 776 ms
Memory Usage: 74.4 MB
"""
from time import time
class LRUCache:
    def __init__(self, capacity: int):
        self.cache = dict()
        self.capacity = capacity

    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        else:
            out = self.cache[key]
            del self.cache[key]
            self.cache[key] = out
            #print('get',self.cache)
            return out

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            del self.cache[key]
        elif len(self.cache) >= self.capacity:
            for k in self.cache.keys():
                first = k
                break
            del self.cache[first]
        self.cache[key] = value
        #print('put',self.cache)
```

```python
"""
Runtime: 3421 ms
Memory Usage: 74.7 MB
"""
from collections import deque
class LRUCache:
    def __init__(self, capacity: int):
        self.cache = dict()
        self.capacity = capacity
        self.history = deque()
        
    def restore(self, key: int):
        del self.history[self.history.index(key)]
        self.history.appendleft(key)
        
    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        else:
            self.restore(key)
            return self.cache[key]

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self.restore(key)
        elif len(self.history) >= self.capacity:
            evict = self.history.pop()
            del self.cache[evict]
            self.history.appendleft(key)
        else:
            self.history.appendleft(key)
        self.cache[key] = value
```

```python
"""
Runtime: 6355 ms
Memory Usage: 75.5 MB
"""
from time import time
class LRUCache:
    def __init__(self, capacity: int):
        self.cache = dict()
        self.capacity = capacity
        self.history = dict()

    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        else:
            self.history[key] = time()
            return self.cache[key]

    def put(self, key: int, value: int) -> None:
        self.cache[key] = value
        self.history[key] = time()
        if len(self.cache) > self.capacity:
            evict = min(self.history, key=self.history.get)
            del self.cache[evict]
            del self.history[evict]
```
