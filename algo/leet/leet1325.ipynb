<h2>1325. Delete Leaves With a Given Value</h2>
<div><p>Given a binary tree <code>root</code> and an integer <code>target</code>, delete all the <strong>leaf nodes</strong> with value <code>target</code>.</p>

<p>Note that once you delete a leaf node with value <code>target</code><strong>, </strong>if its parent node becomes a leaf node and has the value <code>target</code>, it should also be deleted (you need to continue doing that until you cannot).</p>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li>The number of nodes in the tree is in the range <code>[1, 3000]</code>.</li>
	<li><code>1 &lt;= Node.val, target &lt;= 1000</code></li>
</ul>
</div>

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
# 47 ms	14.7 MB
class Solution:
    def removeLeafNodes(self, root: Optional[TreeNode], target: int) -> Optional[TreeNode]:
        if self.DFS(root,target):
            return None
        return root
    def DFS(self, node, target):
        if not node:
            return True
        left_none = self.DFS(node.left,target)
        right_none = self.DFS(node.right,target)
        if left_none:
            node.left = None
        if right_none:
            node.right = None
        if left_none and right_none and node.val == target:
            return True
        else:
            return False
```
